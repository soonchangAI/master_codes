```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
import keras
np.random.seed(0)


training_starts = np.load('training_starts.npy')
training_durations = np.load('training_durations.npy')
starts = np.load('normal_starts.npy')
abnormal_starts = np.load('abnormal_starts.npy')
durations = np.load('durations.npy')

print(training_starts.shape)
print(training_durations.shape)
print(starts.shape)
print(abnormal_starts.shape)
print(durations.shape)
```

```python
training_data = np.vstack([training_starts,training_durations]).T
print(training_data.shape)
test_normal_data = np.vstack([starts,durations]).T
print(test_normal_data.shape)
test_abnormal_data = np.vstack([abnormal_starts,durations]).T
print(test_abnormal_data.shape)
```

```python
from tensorflow import set_random_seed
set_random_seed(0)
model = Sequential()
model.add(Dense(4,input_dim=2,activation='relu',kernel_initializer=keras.initializers.RandomNormal(seed=0)))
model.add(Dense(6,activation='relu',kernel_initializer=keras.initializers.RandomNormal(seed=0)))
model.add(Dense(8,activation = 'relu',kernel_initializer=keras.initializers.RandomNormal(seed=0)))
model.add(Dense(6,activation='relu',kernel_initializer=keras.initializers.RandomNormal(seed=0)))
model.add(Dense(2,kernel_initializer=keras.initializers.RandomNormal(seed=0)))

model.compile(loss='mse',metrics=['mse'],optimizer='adam')
model.fit(training_data,training_data,epochs=500)
```

```python
model.evaluate(test_abnormal_data,test_abnormal_data)
```

```python
correct_normal = 0
threshold = 22301.4512
for i in range(len(test_normal_data)):
    test = test_normal_data[i,:].reshape(1,2)
    loss, mse = model.evaluate(test,test)
    if loss <= threshold:
        correct_normal += 1
    
print(correct_normal)
```

```python
correct_abnormal = 0
for i in range(len(test_abnormal_data)):
    test = test_abnormal_data[i,:].reshape(1,2)
    loss, mse = model.evaluate(test,test)
    if loss > threshold:
        correct_abnormal += 1
    
print(correct_abnormal)
```
